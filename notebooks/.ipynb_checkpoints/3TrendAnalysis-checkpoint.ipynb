# Analyzing Trend Following Returns

```python
import os
import sys
import numpy as np
import pandas as pd
import datetime as dt
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

# Preparing data

```python
fut_path    = r"C:\Users\Diego\Documents\GitHub\TrendConvexity\data\FuturesData.parquet"
hf_path     = r"C:\Users\Diego\Documents\GitHub\TrendConvexity\data\hf_indices.parquet"
ticker_path = r"A:\BBGFuturesManager_backup_backup\root\fut_tickers.xlsx"
spx_path    = r"C:\Users\Diego\Documents\GitHub\TrendConvexity\data\spx.parquet"
```

```python
df_px = (pd.read_parquet(
    path = fut_path, engine = "pyarrow").
    pivot(index = "date", columns = "security", values = "PX_LAST"))

df_ticker_guide = (pd.read_excel(
    io = ticker_path).
    assign(tmp = lambda x: 
           np.where(x.contract.str.len() == 1, x.contract + " 1", x.contract + "1") + " " + x.ending)
    [["tmp", "name", "kind"]].
    rename(columns = {
        "tmp" : "ticker",
        "kind": "Asset Class"}))

df_cta = (pd.read_parquet(
    path = hf_path, engine = "pyarrow").
    query("security == 'NEIXCTA Index'").
    set_index("date")
    [["PX_LAST"]].
    pct_change().
    dropna())

df_spx = (pd.read_parquet(
    path = spx_path, engine = "pyarrow").
    query("security == 'SPX Index'").
    rename(columns = {"PX_LAST": "SPX"}).
    set_index("date")
    [["SPX"]])
```

```python
df_signal = (df_px.diff().apply(
    lambda x: x / x.ewm(span = 10, adjust = False).std()).
    apply(lambda x: x.ewm(span = 180, adjust = False).mean()).
    shift().
    apply(lambda x: np.sign(x)))
```

```python
df_signal_rtn = df_px.pct_change() * df_signal
df_weight     = (df_signal_rtn.apply(
    lambda x: 0.01 / (x.ewm(span = 100, adjust = False).std() * np.sqrt(252))))
```

```python
df_rtn = (df_px.pct_change().reset_index().melt(
    id_vars = "date", value_name = "px_rtn").
    dropna())
```

```python
df_combined = ((df_signal_rtn * df_weight).reset_index().melt(
    id_vars = "date", value_name = "tf_rtn").
    dropna().
    merge(right = df_rtn, how = "inner", on = ["date", "security"]))
```

# Comparing convexity and volatility compared against returns

```python
def _get_convexity(df: pd.DataFrame) -> pd.DataFrame: 
    
    model = (sm.OLS(
        endog = df.tf_rtn,
        exog  = sm.add_constant(df[["px_rtn", "sq_rtn"]])).
        fit())
    
    df_param = (model.params.to_frame(
        name = "param_val").
        reset_index())

    df_value = (model.pvalues.to_frame(
        name = "pvalue").
        reset_index())
    
    df_tstats = (model.tvalues.to_frame(
        name = "t_stat").
        reset_index())
    
    df_out = (df_param.merge(
        right = df_value, how = "inner", on = ["index"]).
        merge(right = df_tstats, how = "inner", on = ["index"]))
    
    return df_out
    
df_ols_params = (df_combined.assign(
    sq_rtn = lambda x: x.px_rtn ** 2).
    groupby("security").
    apply(_get_convexity).
    reset_index().
    drop(columns = ["level_1"]).
    rename(columns = {"index": "param"}))
```

```python
df_stats = (df_combined.melt(
    id_vars = ["date", "security"]).
    drop(columns = ["date"]).
    groupby(["security", "variable"]).
    agg(["mean", "std"])
    ["value"].
    rename(columns = {
        "mean": "mean_rtn",
        "std" : "std_rtn"}).
    assign(
        sharpe  = lambda x: x.mean_rtn / x.std_rtn * np.sqrt(252),
        ann_std = lambda x: x.std_rtn * np.sqrt(252)).
    reset_index())
```

```python
df_spread = (df_stats.pivot(
    index = "security", columns = "variable", values = "sharpe").
    assign(sharpe_spread = lambda x: x.tf_rtn - x.px_rtn))

df_vol = (df_stats.query(
    "variable == 'px_rtn'")
    [["security", "std_rtn"]])
```

```python
df_combined_ols = (df_spread.merge(
    right = df_vol, how = "inner", on = ["security"]).
    merge(right = df_ols_params, how = "inner", on = ["security"]).
    query("param == 'sq_rtn'"))
```

```python
fig, axes = plt.subplots(ncols = 3, figsize = (20,5))

df_combined_ols.plot(
    kind = "scatter", 
    x    = "std_rtn",
    y    = "sharpe_spread",
    ax   = axes[0],
    xlabel = "Annualized Underlying Volatility",
    ylabel = "TF Sharpe - Underlying Sharpe",
    title  = "Comparing Underlying Volatility")

(df_combined_ols.assign(
    tmp = lambda x: np.abs(x.param_val)).
    query("tmp <= 5").
    plot(
        kind   = "scatter",
        x      = "param_val",
        y      = "sharpe_spread",
        ax     = axes[1],
        ylabel = "TF Sharpe - Underlying Sharpe",
        xlabel = "Convexity from OLS (After Controlling for Beta)",
        title  = "Comparing Convexity from OLS (After Controlling for Beta)\nNot Including STIR Products"))

(df_combined_ols.assign(
    tmp = lambda x: np.abs(x.param_val)).
    query("t_stat <= 30").
    plot(
        kind   = "scatter",
        x      = "t_stat",
        y      = "sharpe_spread",
        ax     = axes[2],
        ylabel = "TF Sharpe - Underlying Sharpe",
        xlabel = "t-stat of Convexity from OLS (After Controlling for Beta)",
        title  = "Comparing t-stat of Convexity from OLS (After Controlling for Beta)\nNot Including STIR Products"))

fig.suptitle("Comparing Convexity Possible Drivers of Trend Following Returns vs. The over/underperformance of Trend Following")
plt.tight_layout()
```
